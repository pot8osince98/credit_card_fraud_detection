```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(class_weight='balanced')
```

```python
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,plot_confusion_matrix,make_scorer,recall_score
```

```python
print(classification_report(y_test,preds))
```

```python
plot_confusion_matrix(model,X_test,y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
scorer = make_scorer(recall_score,pos_label=0)
```

```python
k_model = KNeighborsClassifier()
```

```python
k_model.fit(X_train,y_train)
```

```python
pred = k_model.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```

```python
plot_confusion_matrix(k_model,X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
log_preds = log_model.predict(X_test)
```

```python
print(classification_report(y_test,log_preds))
```

```python
k_values = list(range(1,20))
param_grid = {'n_neighbors': k_values}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(k_model,param_grid,scoring=scorer)
```

```python
grid.fit(X_train,y_train)
```

```python
k_preds = grid.predict(X_test)
```

```python
print(classification_report(y_test,k_preds))
```

```python
plot_confusion_matrix(grid,X_test,y_test)
```

```python
grid = GridSearchCV(k_model,param_grid,scoring='recall')
```

```python
grid.fit(X_train,y_train)
```

```python
k_preds_two = grid.predict(X_test)
```

```python
print(classification_report(y_test,k_preds_two))
```

```python
plot_confusion_matrix(grid,X_test,y_test)
```
